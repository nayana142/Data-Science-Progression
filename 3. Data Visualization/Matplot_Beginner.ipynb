## Line Chart

The line chart is one of the simplest and most widely used data visualization techniques. A line chart displays information as a series of data points or markers connected by straight lines. You can customize the shape, size, color, and other aesthetic elements of the lines and markers for better visual clarity.

Here's a Python list showing the yield of apples (tons per hectare) over six years

```python
yield_apples = [0.85, 0.933, 0.962, 0.976, 0.981, 0.991]
```

# Box-and-whisker

* Another common statistical plot for exploring a numerical feature is a box-and-whisker plot, also referred to as a boxplot.

* Box-and-whisker plots display the distribution of an array based on the minimum, first quartile, median, third quartile, and maximum, but they are primarily used to indicate the skew of a distribution and to identify outliers.

? Generate an array of 100 normally distributed numbers with a mean of 0 and a standard deviation of 0.1.

```python
import numpy as np
y = np.random.normal(loc=0, scale=0.1, size=100)
```

```python
plt.boxplot(y) 
plt.show()
```

* The plot displays a box that represents the interquartile range (IQR).
* The top of the box is the 25th percentile (i.e., Q1) while the bottom of the box is the 75th percentile (that is, Q3). 
* The orange line going through the box is the median.
* The two lines extending above and below the box are the whiskers.
* The top of the upper whisker is the "maximum" value, which is calculated using Q1  1.5*IQR.
* The bottom of the lower whisker is the "minimum" value, which is calculated using Q3 + 1.5*IQR. 
* Outliers (or fringe outliers) are displayed as dots above the "maximum" whisker or below the "minimum" whisker.

```python
# Sample data creation
import pandas as pd
data = {
   
    'windspeed': [10, 15, 8, 20, 12, 0, 7, 14, 21, 18, 9, 6, 3, 11, 5, 13, 17, 19, 22, 16, 8, 10, 7, 5, 13]
}
df = pd.DataFrame(data)
```

* We want our plot title to inform the reader about the strength of the relationship and the Pearson correlation coefficient. Thus, we will calculate the Pearson correlation coefficient and interpret the value of the coefficient in the title. To compute the Pearson correlation coefficient

# Pearson Correlation Coefficient (r):

* The Pearson correlation coefficient is a measure of the linear relationship between two variables. It ranges from -1 to 1:

    * 1: Perfect positive linear relationship.
    * 0: No linear relationship.
    * -1: Perfect negative linear relationship.
    

### Returns
    
* correlation_coefficient: The Pearson correlation coefficient (r value).
* p_value: The two-tailed p-value for testing non-correlation.


### Interpretation
* If r is close to 1 or -1 and the p-value is low (e.g., < 0.05), there is a strong linear relationship between x and y.
* If r is close to 0 and the p-value is high, it indicates little to no linear relationship between x and y.

```python
from scipy.stats import pearsonr
correlation_coeff, p_value = pearsonr(x, y)
```

# Pie Chart

* In a pie chart, each slice stands for a different category, and the size of the slice shows how much of the total each category makes up.

* Consider a company's budget where the spending is divided among different departments
     * Marketing
     * Research & Development
     * Sales, and Operations
* A pie chart could effectively show how much of the total budget each department consumes, allowing stakeholders to visually grasp the relative importance or scale of spending in each area.

```python
# Data to plot
labels = ['Category A', 'Category B', 'Category C', 'Category D']
# The sizes for each slice
sizes = [20, 30, 25, 25] 
# Optional: custom colors
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] 
# Explode the 1st slice 
explode = (0.1, 0, 0, 0)  
```

```python
plt.pie(sizes, explode=(0,0,0.1,0), labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=10);
plt.title('Company Budget Distribution Across Departments');
```

# Explanation:
* labels: A list of labels for each slice of the pie.
* sizes: A list of numerical values that determine the size of each slice. The total of these values is used to determine the proportional size of each slice.
* colors: (Optional) A list of colors for each slice.
* explode: (Optional) A tuple that indicates how much to "explode" or separate each slice from the pie. For example, (0.1, 0, 0, 0) will pull the first slice out by 0.1 units.
* autopct: (Optional) A string format to label the slices with their percentage value (e.g., '%1.1f%%').
* shadow: (Optional) Adds a shadow to the pie chart.
* startangle: (Optional) Rotates the chart to start from a different angle (e.g., 140 degrees).

# Object-Oriented Approach Using Subplots

* In the object-oriented approach, we create a figure object that acts as an empty canvas and then we add a set of axes, or subplots, to it
* Another advantage of using the object-oriented plotting approach is the ability to create multiple subplots on a single figure object.
* In some situations, we want to compare different views of data side by side. We can accomplish this in Matplotlib using subplots

? First, create x, y, and y2 using the following code:

```python
import numpy as np
x = np.linspace(0, 10, 20)
y = x**3
y2 = x**2
```

? Create a figure with two axes (that is, subplots) that are side by side (that is, 1 row 
with 2 columns)

* The fig object is now callable and returns the axis on which we can plot

```python
x
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('x by x Cubed')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')

axes[1].plot(x, y2)
axes[1].set_title('x by x Squared')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')
plt.show()
```

* It looks like the y-axis of the plot on the right is overlapping the left-hand plot. To prevent the overlapping of the plots, use plt.tight_layout().

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('x by x Cubed')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')
axes[1].plot(x, y2)
axes[1].set_title('x by x Squared')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')
plt.tight_layout()
plt.show()
```

# Plots that primarily use numerical values include:

* Histogram: Represents the distribution of numerical data by showing the frequency of data points within specified ranges (bins).

* Box Plot (Box-and-Whisker Plot): Summarizes the distribution of numerical data through their quartiles and outliers.

* Scatter Plot: Displays points based on two numerical variables, allowing you to see correlations or patterns.

* Line Plot: Shows trends over time or continuous data by connecting data points with lines.

* Bar Plot (with numerical data): Can be used when both the categories (on the x-axis) and their corresponding values (on the y-axis) are numerical.

* Heatmap: Represents numerical data in a matrix format with colors indicating the magnitude of values.

* Violin Plot: Combines a box plot with a density plot, showing the distribution of numerical data across different categories.

* Density Plot: Estimates the probability density function of a continuous numerical variable.

* Pair Plot: Displays scatter plots of all possible pairs of numerical variables in a dataset.

# Common plots that can use both categorical and numerical values:

* Bar Plot: Visualizes the relationship between a categorical variable (x-axis) and a numerical variable (y-axis). Each bar represents a category.

* Stacked Bar Plot: Similar to a bar plot but shows multiple categories within each bar, stacked on top of each other, allowing for comparison between different groups.

* Pie Chart: Displays the proportions of different categories within a dataset. Each slice of the pie represents a category.

* Box Plot (Box-and-Whisker Plot): While it primarily deals with numerical data, it can group the data by a categorical variable to compare distributions across categories.

* Violin Plot: Combines a box plot and a density plot, displaying the distribution of numerical data across different categories.

* Swarm Plot: Similar to a strip plot but adjusts points to avoid overlap, showing the distribution of a numerical variable across categories.

* Count Plot: A type of bar plot that counts the occurrences of each category and displays them as bars.

* Point Plot: Similar to a line plot but specifically used for categorical data, it shows the mean value of a numerical variable for each category, with points connected by lines.

* Heatmap: Although primarily used for numerical data, it can also be applied when one or both axes represent categorical variables, showing how numerical values change across categories.
