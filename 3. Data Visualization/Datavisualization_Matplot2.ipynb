# Histplot

* In statistics, it is essential to be aware of the distribution of continuous variables prior to running any type of analysis. To display the distribution, we will use a histogram. 
* Histograms display the frequency by the bin for a given array

```python
import random
import pandas as pd
import numpy as np
np.random.seed(42)  # For reproducibility
data = {
    'Age': np.random.randint(20, 60, size=100),  # Random ages between 20 and 60
    'Salary': np.random.randint(30000, 100000, size=100),  # Random salaries
    'YearsAtCompany': np.random.randint(1, 20, size=100)  # Random years at the company
}
```

```python
data=pd.DataFrame(data)
```

```python
data
```

```python
plt.title('Salary Distribution')
plt.hist(data['Salary'],bins=5,edgecolor='black');
plt.xlabel('Salary')
plt.show()
```

```python
import numpy as np
import pandas as pd
np.random.seed(42)

data={
    'F1':np.random.normal(loc=50,scale=10,size=500),
    'F2':np.random.normal(loc=55,scale=10,size=500),
    'F3':np.random.normal(loc=60,scale=15,size=500),
    'F4':np.random.normal(loc=70,scale=5,size=500)
}
```

```python
data=pd.DataFrame(data)
```

```python
data
```

```python
data.hist(bins=10,edgecolor='black',color='red')
plt.grid(False)
plt.suptitle('Histplot of data');
```

```python
# random.randint--> integer
# random.rand--> float
# random.normal --> normal distributed points
```

# Pie Chart

```python
data={
'Units_Sold1': [25, 40, 45, 50, 40, 25, 55],
'Product':['Product A', 'Product B', 'Product D', 'Product A', 'Product F', 'Product D', 'Product E']
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.groupby(['Product'])['Units_Sold1'].sum()
```

```python
df.groupby(['Product'])['Units_Sold1'].sum().plot(kind='pie',explode=(0,0,0,0,0.1),shadow=True,autopct="%1.2f%%");
```

```python
# Data to plot
labels = ['Category A', 'Category B', 'Category C', 'Category D']
# The sizes for each slice
sizes = [20, 30, 25, 25]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] 
```

```python
plt.pie(sizes,labels=labels,explode=(0.2,0,0,0),shadow=True,autopct="%1.f%%",startangle=90,colors=colors);
plt.title('pie chart');
```

# Object-Oriented Approach Using Subplots

```python
import numpy as np
x = np.linspace(0, 10, 20)
x
```

```python
y = x**3
y2 = x**2
```

```python
y
```

```python
y2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('x by x Cubed')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')

axes[1].plot(x, y2)
axes[1].set_title('x by x Squared')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')
plt.show()
```
