### How many mid-level (MI) data engineers are working in the UK with a salary between 80,000 and 100,000 USD?

```python
DE=df[(df['experience_level']=='MI') & (df['job_title']=='Data Engineer') & (df['salary_in_usd']>=80000) & (df['salary_in_usd']>=100000)].reset_index(drop=True)
DE
```

### How does salary in USD vary with experience level (junior,Â Mid,Â Senior)?

```python
df1=df.groupby(['experience_level','job_title'])['salary_in_usd'].mean().to_frame().reset_index()
df1
```

### salary changes across different experience levels.

```python
plt.figure(figsize=(10,6))
plt.title("salary changes across different experience levels",color='red')
plt.bar(df1['experience_level'],df1['salary_in_usd'],color='grey')
plt.xlabel("experience_level")
plt.ylabel("average salary_in_usd")           
plt.show()
```

### What is the average remote work ratio for companies of different sizes and locations?

```python
df2=df.groupby(['company_size','company_location'])['remote_ratio'].mean().to_frame().reset_index()
df2
```

### Compare the remote work flexibility offered by companies of different sizes.

```python
plt.figure(figsize=(10,5))
plt.title('Remote Ratio Distribution by Company Size')
plt.hist(df2['remote_ratio'],bins=10,color='pink',edgecolor='black')
plt.xlabel('remote_ratio')
plt.ylabel('frequency')
;
```

### Group by employment_type and plot the average salary_in_usd for each type.

```python
df3=df.groupby('employment_type')['salary_in_usd'].mean().to_frame().reset_index()
df3
```

### Compare salaries across different types of employment

```python
plt.title('Salary Distribution by Employment Type')
plt.barh(df3['employment_type'],df3['salary_in_usd'],color='skyblue')
plt.xlabel('employment_type')
plt.ylabel('avg salary_in_usd')
plt.show()
```
