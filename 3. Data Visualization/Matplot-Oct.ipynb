```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
y=[160,173,175,179,180]
# y=['std1','std2','std3','std4','std5']
x=[2000,2001,2002,2003,2004]
```

```python
plt.figure(figsize=(8,6))
plt.title('height of student')
plt.plot(x,y,marker='*')
plt.xlabel('year')
plt.ylabel('height')
plt.show()
```

```python
tshirt=['adidas','puma','nike','levis']
price=[1500,1800,2000,1400]
```

```python
plt.figure(figsize=(8,5))
plt.title('price comparison of tshirts')
plt.bar(tshirt,price)
plt.xlabel('brands')
plt.ylabel('price')
plt.show()
```

```python
data={'tshirt':['adidas','puma','nike','levis'],
'price':[1500,1800,2000,1400]}
df=pd.DataFrame(data)
df.set_index('tshirt',inplace=True)
```

```python
df
```

```python
df.plot(kind='line')
plt.show()
```

```python
df.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Price comparison of tshirts')
plt.bar(df['tshirt'],df['price'])
plt.xlabel('brands')
plt.ylabel('price')
plt.show()
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Sedans': [150, 200, 180, 220],
    'SUVs': [120, 180, 160, 200],
    'Trucks': [80, 100, 90, 120]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.set_index('Region',inplace=True)
```

```python
df
```

```python
df.plot(kind='bar')
plt.title('unit sold across different region')
plt.legend(title='Type of vehicle')
plt.show()
```

```python
df.plot(kind='bar',stacked=True)
plt.title('unit sold across different region')
plt.legend(title='Type of vehicle')
plt.show()
```
