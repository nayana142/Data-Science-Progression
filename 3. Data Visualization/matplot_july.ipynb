```python
# multiple plots
sales2 = [65000, 74000, 79000, 84000, 87000, 90000, 95000, 97000, 100000, 106000, 130000, 145000]
sales = [75000, 78000, 82000, 85000, 87000, 90000, 93000, 95000, 100000, 105000, 110000, 115000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.figure(figsize=(6,5))
plt.plot(months,sales,marker='*',color='g')
plt.plot(months,sales2,marker='s',color='red')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.title('sales over month');
```

```python
plt.figure(figsize=(6,5))
plt.plot(months,sales,marker='*',color='g')
plt.plot(months,sales2,marker='s',color='red')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend(['C1','C2'])
plt.xticks(rotation=90)
plt.title('sales over month');
```

# Histplot

* A histogram (or hist plot) is a graphical representation of the distribution of numerical data. 
* It shows the frequency of data points falling within specified ranges, called bins or intervals.
* The width of each bin corresponds to the range of values it covers.
* The height of each bar represents the frequency (or count) of data points within the bin’s range.
* A taller bar indicates a higher frequency of data points within that range

```python
import random
import pandas as pd
import numpy as np
np.random.seed(42)  # For reproducibility
data = {
    'Age': np.random.randint(20, 60, size=100),  # Random ages between 20 and 60
    'Salary': np.random.randint(30000, 100000, size=100),  # Random salaries
    'YearsAtCompany': np.random.randint(1, 20, size=100)  # Random years at the company
}
```

```python
data=pd.DataFrame(data)
```

```python
data
```

```python
plt.title('Salary Distribution')
plt.hist(data['Salary'],bins=5,edgecolor='black');
plt.xlabel('Salary')
plt.show()
```
