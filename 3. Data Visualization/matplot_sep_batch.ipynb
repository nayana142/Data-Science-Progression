```python
# Compairing two companies profit
```

```python
profit=[200,250,150,300,290,350]
profit2=[210,290,340,270,390,400]
year=[2000,2001,2002,2003,2004,2005]
```

```python
plt.figure(figsize=(8,7))
plt.title('Compairing profit')
plt.plot(year,profit,marker="*",color="green")
plt.plot(year,profit2,marker="d",color="red")
plt.xlabel('year')
plt.ylabel('profit')
plt.legend(['C1','C2'])
plt.xticks(rotation=90)
plt.show()
```

# Histogram

* In statistics, it is essential to be aware of the distribution of continuous variables prior to running any type of analysis. To display the distribution, we will use a histogram. 

```python
import numpy as np
import pandas as pd
import random
np.random.seed(42)
data = {
    'Age': np.random.randint(20, 60, size=100),  # Random ages between 20 and 60
    'Salary': np.random.randint(30000, 100000, size=100),  # Random salaries
    'YearsAtCompany': np.random.randint(1, 20, size=100)  # Random years at the company
}
```

```python
data=pd.DataFrame(data)
```

```python
data
```

```python
plt.hist(data.Salary,bins=10,edgecolor='black')
plt.show()
```

```python
# normal distributon
np.random.seed(42)
data = {
    'Feature A': np.random.normal(loc=50, scale=10, size=1000),  
    'Feature B': np.random.normal(loc=60, scale=15, size=1000),  
    'Feature C': np.random.normal(loc=55, scale=5, size=1000) ,  
    'Feature D': np.random.normal(loc=70, scale=15, size=1000)    
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
plt.hist(df['Feature A'],bins=10,edgecolor='black')
plt.grid(True);
```

```python
df.hist(bins=10,edgecolor='black',grid=False);
```

# Boxplot

```python
import numpy as np
import pandas as pd
np.random.seed(42)
data = {
    'Category A': np.random.normal(50, 10, 100),  
    'Category B': np.random.normal(60, 20, 100),  
    'Category C': np.random.normal(55, 5, 100),   
    'Category D': np.append(np.random.normal(45, 10, 95), [100, 105, 110, 115, 120])  
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
plt.title('Boxplot of Category A')
plt.boxplot(df['Category A'])
plt.show()
```

```python
df.describe()
```

```python
df.boxplot(labels=df.columns)
```

# Violin plot
* A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. 
* It is used to visualize the distribution of numerical data.
* Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
plt.violinplot(df['Category A']);
```
