# Histogram

```python
data=pd.DataFrame(data)
```

# Boxplot

* A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). 
* It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical how tightly your data is grouped, and if and how your data is skewed.
* The IQR (the range between Q1 and Q3) forms the box, showing the spread of the middle 50% of the data.
* A smaller IQR indicates less variability in the central data, while a larger IQR suggests more spread in the middle values.

```python
import matplotlib.pyplot as plt
windspeed= [10, 15, 8, 20, 12, 0, 7, 14, 21, 18, 9, 6, 3, 11, 5, 13, 17, 19, 22, 16, 8, 10, 7, 5, 13]
plt.boxplot(windspeed);
```

```python
import numpy as np
import pandas as pd
np.random.seed(42)
data = {
    'Category A': np.random.normal(50, 10, 100),  # Normal distribution, mean=50, std=10
    'Category B': np.random.normal(60, 20, 100),  # Normal distribution, mean=60, std=20
    'Category C': np.random.normal(55, 5, 100),   # Normal distribution, mean=55, std=5
    'Category D': np.append(np.random.normal(45, 10, 95), [100, 105, 110, 115, 120])  # Skewed with outliers
}
data=pd.DataFrame(data)
```

```python
data
```

```python
plt.boxplot([data['Category A'],data['Category B']],labels=['A','B'])
plt.legend(['CatgoryA','category B']);
```

# Violin plot
* A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. 
* It is used to visualize the distribution of numerical data.
* Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
plt.violinplot(windspeed);
plt.xlabel('wind')
```
