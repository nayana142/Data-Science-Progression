# Data Description of ds_salaries dataset

* work_year: The year in which the work took place or was recorded.

* experience_level: The level of experience of the employee (e.g., Junior, Mid-level, Senior).

* employment_type: The type of employment (e.g., Full-time, Part-time, Contract).

* job_title: The title of the job (e.g., Data Analyst, Data Scientist).

* salary: The salary received by the employee.

* salary_currency: The currency in which the salary is paid.

* salary_in_usd: The salary converted to USD for standardization.

* employee_residence: The location of the employee's residence.

* remote_ratio: The percentage of the job that can be done remotely.

* company_location: The location of the company.

* company_size: The size of the company (e.g., Small, Medium, Large).

### Q) How many Senior-level Data Analysts are employed full-time?

```python
da=df[(df['experience_level']=='SE')&(df['employment_type']=='FT')&(df['job_title']=='Data Analyst')].reset_index(drop=True)
```

```python
da
```

### Q) Filter employees from large companies who work at least 50% remotely.

```python
remote=df[(df['company_size']=='L')&(df['remote_ratio']>=50)].reset_index(drop=True)
```

```python
remote
```

### Q) How many mid-level (MI) data engineers are working in the UK with a salary between 80,000 and 100,000 USD?

```python
de=df[(df['experience_level']=='MI')&(df['job_title']=='Data Engineer')&(df['salary_in_usd']>=80000)&(df['salary_in_usd']<=100000)].reset_index(drop=True)
```

```python
de
```

### Q) How does salary in USD vary with experience level (junior, Mid, Senior)?

```python
experience_salary=df.groupby(['experience_level','job_title'])['salary_in_usd'].mean().to_frame().reset_index()
```

```python
experience_salary
```

```python
plt.bar(experience_salary['experience_level'],experience_salary['salary_in_usd'],color='pink')
plt.title('Salary Changes Across Different Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
```

### Q) What is the average remote work ratio for companies of different sizes and locations?

```python
avg_remote=df.groupby(['company_location','company_size'])['remote_ratio'].mean().reset_index()
```

```python
avg_remote
```

```python
plt.hist(avg_remote['remote_ratio'],bins=10,color='skyblue',edgecolor='grey')
plt.title('remote work flexibility')
plt.xlabel('Remote Ratio')
plt.ylabel('Frequency')
plt.show()
```

### Q) Group by employment_type and plot the average salary_in_usd for each type.

```python
avg_slry=df.groupby('employment_type')['salary_in_usd'].mean().to_frame().reset_index()
```

```python
avg_slry
```

```python
plt.figure(figsize=(10,5))
plt.title('Salary Distribution by Emplyment',color='g')
plt.barh(avg_slry['employment_type'],avg_slry['salary_in_usd'],color='black')
plt.xlabel('Salary in USD')
plt.ylabel('Emplyment Typre')
plt.show()
```
