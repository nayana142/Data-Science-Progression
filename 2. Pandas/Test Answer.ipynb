# Pandas Task 01

```python
import pandas as pd
```

```python
data = { 'Transaction ID': range(1, 51),
    'Date': pd.date_range(start='2024-01-01', periods=50, freq='D'),
    'Product Category': ['Electronics', 'Furniture', 'Clothing', 'Groceries', 'Books'] * 10,
    'Product Name': ['Laptop', 'Chair', 'T-Shirt', 'Milk', 'Novel'] * 10,
    'Units Sold': [10, 5, 20, 15, 7] * 10,
    'Unit Price': [1000, 200, 20, 2, 15] * 10,
    'Total Revenue': [10000, 1000, 400, 30, 105] * 10,
    'Region': ['North', 'South', 'East', 'West', 'Central'] * 10,
    'Payment Method': ['Credit Card', 'Debit Card', 'Cash', 'Credit Card', 'Cash'] * 10  }
df=pd.DataFrame(data)
df
```

# Q.6:
* convert the dataset from wide format to long format with Transaction ID, Date, Product Category, Product Name, Region, and Payment Method as identifier variables, and Units Sold, Unit Price, and Total Revenue as measured variables.

```python
df_melt=pd.melt(df,id_vars=['Transaction ID','Date','Product Category','Product Name','Region','Payment Method'],value_vars=['Units Sold','Unit Price','Total Revenue'])
df_melt
```

# Q.7:
* Create a pivot table to summarize total revenue for each product category by region.

```python
# Pivot the DataFrame to create a summary table
df_pivot=df.pivot_table(index='Region',columns=['Product Category'],values='Total Revenue',aggfunc='sum')
df_pivot
```

# Q.8:

```python
df_pivot=df.pivot_table(index='Date',columns=['Product Category'],values='Units Sold',aggfunc='sum')
df_pivot
```

# Q.13:
* Use the apply function to create a new column called Discounted Revenue which is 90% of the Total Revenue

```python
def discount(df):
    return df['Total Revenue']*(90/100)
```

```python
df['Discounted Revenue']=df.apply(discount,axis=1)
```
