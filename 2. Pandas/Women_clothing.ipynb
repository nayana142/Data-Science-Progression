## Data Description
* order_id: Unique identifier for each order (integer). 
* order_date: The date on which the order was placed (string). 
* sku: Stock Keeping Unit, which uniquely identifies each product or item (string). 
* color: The color attribute of the product (string). 
* size: The size of the product (string).  
* unit_price: The price per unit of the product (integer). 
* quantity: The number of units ordered (integer). 
* revenue: The total revenue generated from the order (integer).
